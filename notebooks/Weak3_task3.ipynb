```python
import pandas as pd

file_path = '/content/drive/MyDrive/MachineLearningRating_v3.csv'
df = pd.read_csv(file_path, delimiter='|')  # Use '|' as delimiter
df.info()
df.describe()
```

```python
print(df.columns)
```

```python
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
import matplotlib.pyplot as plt

numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

for col in numerical_columns:
    plt.figure(figsize=(8, 5))
    plt.hist(df[col], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
plt.figure(figsize=(8, 5))
# Check if the column name is 'TotalClaims' instead of 'TotalClaim'
plt.scatter(df['TotalPremium'], df['TotalClaims'], alpha=0.5)
plt.title('Total Premium vs Total Claim')
plt.xlabel('Total Premium')
plt.ylabel('Total Claim')
plt.show()
```

```python
for col in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```
